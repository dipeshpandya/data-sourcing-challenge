### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()
tmdb_api_key = ''
nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
tmdb_key_string = "&api_key=" + tmdb_api_key
```

### Access the New York Times API

```python
# Set the base URL
ny_url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"
#page=1
# Build URL
nyt_query_url = ny_url + 'fl='+field_list+'&fq=' + filter_query + '&begin_date=' + begin_date + '&end_date=' + end_date +'&sort='+sort+'&api-key='+nyt_api_key


```

```python
# Create an empty list to store the reviews
review_list = []
        

#print(json.dumps(review_list, indent=4))
#articles = requests.get(reviews).json()
nyt_page = 0

pages_less_than = 20     # # of pages to loop through. 

# loop through pages 0-19
for nyt_page in range(pages_less_than):


    # create query with a page number
    # API results show 10 articles at a time
    page_query_url = f"{nyt_query_url}&page={str(nyt_page)}"

    
    # Make a "GET" request and retrieve the JSON
    articles = requests.get(page_query_url).json()

    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list

        # loop through the reviews["response"]["docs"] and append each review to the list
    try:
        for review in articles["response"]["docs"]:
#        reviews = articles["response"]["docs"]
            review_list.append(review)

        # Print the page that was just retrieved
        print(f'Found Page # - {nyt_page}')
    except:
        print(f'Page # - {nyt_page} not found')

        # Print the page number that had no results then break from the loop
    
```

```python
reviewlist_df = pd.DataFrame(review_list)
reviewlist_df
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
normalized_review_list = pd.json_normalize(review_list, max_level=1)
normalized_review_list
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
review_list_df = pd.DataFrame(normalized_review_list)
review_list_df.head()
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
review_list_df['title'] = review_list_df['headline.main'].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])
review_list_df.head()
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords



# Fix the "keywords" column by converting cells from a list to a string
review_list_df['keywords'] = review_list_df['keywords'].apply(extract_keywords)
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database

titles = review_list_df['title'].to_list()
```

### Access The Movie Database API

```python
# Create an empty list to store the results
tmdb_movies_list=[]

tmdb_url = 'https://api.themoviedb.org/3/search/movie?query='

tmdb_key_string = "&api_key=" + tmdb_api_key

sl_key = 'english_name' # variable to use for languages keys
gpc_key = 'name' #key to use for genres and production countries list

full_moview_url = 'https://api.themoviedb.org/3/movie/'
request_counter = 1 # counter to check total # of api calls
req_number = 1 # this variable is to count total # of movies requested

# Create a request counter to sleep the requests after a multiple
# of 50 requests
max_range = 50

# Loop through the titles

for movie in titles:
    # Check if we need to sleep before making a request
    
    if request_counter == 51:
        
        time.sleep(1)
        print('Taking a quick break...')
    #else:
    # Add 1 to the request counter
        request_counter = 1

    # Make a request to search for movie title
    tmdb_query_url = f"{tmdb_url}{movie}&{tmdb_key_string}"
    # Perform a "GET" request for The Movie Database

    tmdb_movie = requests.get(tmdb_query_url).json()

    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.

    try:
        tmdb_movie['total_results']

        # Get movie id        
        movieid = tmdb_movie['results'][0]['id']

        # Make a request for a the full movie details
        tmdb_full_url = f'{full_moview_url}{movieid}?{tmdb_key_string}'

        # Execute "GET" request with url
        full_detail1 = requests.get(tmdb_full_url).json()
        print(f'Found {req_number} - {movie}')


        req_number +=1 # this variable is to count total # of movies
        request_counter += 1
        sl_length = len(full_detail1['spoken_languages'])
        pc_length = len(full_detail1['production_countries'])
        genres_lenght = len(full_detail1['genres'])
                        #        print(f"{movieid} - {length}")
        spoken_languages=[]
        genres=[]
        production_countries=[]

        # Extract the spoken_languages' English name into a list
        for i in range(sl_length):
            spoken_languages.append(full_detail1['spoken_languages'][i][sl_key])
            i=i+1

        # Extract the genre names into a list
        for i in range(genres_lenght):
            genres.append(full_detail1['genres'][i][gpc_key])
            i=i+1

        # Extract the production_countries' name into a list
        for i in range(pc_length):
            production_countries.append(full_detail1['production_countries'][i][gpc_key])
            i=i+1

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
                        
        tmdb_movies_list.append({
        'title': full_detail1['title']
        ,'original_title': full_detail1['original_title']
        ,'budget': full_detail1['budget']
        ,'original_language' : full_detail1['original_language']
        ,'homepage':full_detail1['homepage']
        ,'overview':full_detail1['overview']
        ,'popularity':full_detail1['popularity']
        ,'runtime':full_detail1['runtime']
        ,'revenue':full_detail1['revenue']
        ,'release_date': full_detail1['release_date']
        ,'vote_average':full_detail1['vote_average']
        ,'vote_count':full_detail1['vote_count']
        ,'genre':genres
        ,'spoken_languages':spoken_languages
        ,'production_countries':production_countries
        })

        # Print out the title that was found

    except:
        print(f'Not found #{req_number} - {movie}')
        req_number +=1

        request_counter += 1
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
print(json.dumps(tmdb_movies_list[:5],indent=4))
```

```python
# Convert the results to a DataFrame
tmdf_df = pd.DataFrame(tmdb_movies_list)
tmdf_df
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merge_df = pd.merge(tmdf_df, review_list_df, on="title")
merge_df
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix=['genre', 'spoken_languages','production_countries']

# Create a list of characters to remove
char_to_replace = ["[", "]", "'"]

# Loop through the list of columns to fix
for c in columns_to_fix:
    
    # Convert the column to type 'str'
    merge_df[c] = merge_df[c].astype('str')

    # Loop through characters to remove
    for i in char_to_replace:
        merge_df[c] = merge_df[c].str.replace(i,'')
    #merge_df[c] = merge_df[c].apply(lambda x: ''.join([x.replace(i, '') for i in char_to_replace]))
# Display the fixed DataFrame
merge_df.head()
```

```python
# Drop "byline.person" column
merge_df = merge_df.drop('byline.person', axis=1)
merge_df.dtypes
```

```python
# Delete duplicate rows and reset index
#merge_df.drop_duplicates(inplace=False)
merge_df = merge_df.drop_duplicates().reset_index(drop=True)
merge_df
```

```python
# Export data to CSV without the index
merge_df.to_csv('output\collected_data_dp.csv',  index=False)
```
